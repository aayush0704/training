```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('https://raw.githubusercontent.com/rajeevratan84/datascienceforbusiness/master/winequalityN.csv')
```

```python
df.head()
```

```python
pop=df['alcohol']
sample_size=500
sample=df['alcohol'].sample(sample_size)
sample_mean=sample.mean()
pop_mean=pop.mean()
error=pop_mean-sample_mean
print(error,pop_mean)
```

```python
plt.scatter(range(1,sample_size+1),sample)
plt.axhline(pop_mean,color='red',label='population mean')
plt.axhline(pop_mean,color='blue',label='sample mean')
plt.legend()
plt.show()
```

```python
def analyze_wine(df, wine_type):
    wine_df = df[df['type'] == wine_type]

    pop = wine_df['alcohol']
    sample_size = 500


    if sample_size > len(pop):
        sample_size = len(pop)

    sample = pop.sample(sample_size, random_state=42)

    sample_mean = sample.mean()
    pop_mean = pop.mean()
    error = pop_mean - sample_mean

    print(f"\n{wine_type.upper()} WINE")
    print("Population mean :", pop_mean)
    print("Sample mean     :", sample_mean)
    print("Error           :", error)

    plt.figure(figsize=(10, 5))
    plt.scatter(range(1, sample_size + 1), sample, alpha=0.6)
    plt.axhline(pop_mean, color='red', label='Population Mean')
    plt.axhline(sample_mean, color='blue', label='Sample Mean')
    plt.xlabel("Sample Index")
    plt.ylabel("Alcohol")
    plt.title(f"{wine_type.capitalize()} Wine: Sample vs Population Mean")
    plt.legend()
    plt.show()
```

```python
analyze_wine(df, 'white')
```

```python
analyze_wine(df, 'red')
```

```python
white_alcohol = df[df['type'] == 'white']['alcohol'].reset_index(drop=True)

plt.figure(figsize=(10,5))
plt.plot(range(len(white_alcohol)), white_alcohol, color='blue')
plt.xlabel("Index")
plt.ylabel("Alcohol")
plt.title("Line Chart of Alcohol Content (White Wine)")
plt.show()
```

```python
red_alcohol = df[df['type'] == 'red']['alcohol'].reset_index(drop=True)

plt.figure(figsize=(10,5))
plt.plot(range(len(red_alcohol)), red_alcohol, color='maroon')
plt.xlabel("Index")
plt.ylabel("Alcohol")
plt.title("Line Chart of Alcohol Content (Red Wine)")
plt.show()
```

```python
white_mean = df[df['type'] == 'white']['alcohol'].mean()
red_mean = df[df['type'] == 'red']['alcohol'].mean()

types = ['White Wine', 'Red Wine']
means = [white_mean, red_mean]

plt.figure(figsize=(7,5))
bars = plt.bar(types, means)

plt.ylabel("Average Alcohol")
plt.title("Average Alcohol Content by Wine Type")


for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2,
             height,
             f"{height:.2f}",
             ha='center',
             va='bottom')

plt.show()
```

```python
type_counts = df['type'].value_counts()
colors = ['#ff9999', '#66b3ff']
plt.figure(figsize=(6,6))
plt.pie(type_counts, labels=type_counts.index, autopct='%1.1f%%',colors=colors)
plt.title("Distribution of Wine Types")
plt.show()
```

```python
white = df[df['type']=='white']['alcohol']
red = df[df['type']=='red']['alcohol']

plt.figure(figsize=(7,5))
plt.boxplot([white, red], labels=['White', 'Red'])
plt.ylabel("Alcohol")
plt.title("Box Plot of Alcohol Content by Wine Type")
plt.show()
```

```python
plt.figure(figsize=(7,5))
plt.violinplot([white, red], showmeans=True)
plt.xticks([1,2], ['White', 'Red'])
plt.ylabel("Alcohol")
plt.title("Violin Plot of Alcohol Content")
plt.show()
```

```python
white_sorted = np.sort(white)

plt.figure(figsize=(8,5))
plt.fill_between(range(len(white_sorted)), white_sorted, alpha=0.6)
plt.xlabel("Index")
plt.ylabel("Alcohol")
plt.title("Area Chart of Alcohol Content (White Wine)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.step(range(len(red)), red.sort_values().values)
plt.xlabel("Index")
plt.ylabel("Alcohol")
plt.title("Step Plot of Alcohol Values (Red Wine)")
plt.show()
```
